# Data Exploration

```python
import pandas as pd
import numpy as np
```

```python
df_full = pd.read_csv('https://raw.githubusercontent.com/brendad8/Datasets/main/t5_leagues.csv')
```

## Shooting Stats

Relationship between Expected Goals and Goals Scored

Above line --> Overperformance
Below line --> Underperformance

```python
df_full.plot.scatter(x='shoot.xG', y='shoot.Gls')
pd.Series(np.linspace(0, 35, 36)).plot.line()
```

Biggest Underperformers

```python
df_full.sort_values('shoot.G-xG', ascending=True)[['Player', 'Pos', 'Squad', 'shoot.Gls', 'shoot.xG', 'shoot.G-xG']].head(10)
```

Biggest Overperformers

```python
df_full.sort_values('shoot.G-xG', ascending=False)[['Player', 'Pos', 'Squad', 'shoot.Gls', 'shoot.xG', 'shoot.G-xG']].head(10)
```

## Defensive Stats

Most Errors leading to shots

```python
df_full.sort_values('defense.Err', ascending=False)[['Player', 'Pos', 'Squad', '90s', 'defense.Tkl+Int', 'defense.Err']].head(10)
```

Dayot Upamecano made the most errors but we can see he does a high volume of defending from his tackle and interception numbers. We also see a lot of goalies which makes sense since they are so close to their own goal. Most errors a goalie makes lead to a shot where this is less true for those higher up the field.  

Offensive Players who defend high up the pitch

```python
df_full.sort_values('defense.Att 3rd', ascending=False)[['Player', 'Pos', 'Squad', '90s', 'defense.Tkl+Int', 'defense.Att 3rd']].head(10)
```
